```python
import pandas as pd
```

```python
df_trips = pd.read_parquet("data/green_tripdata_2025-11.parquet")
```

```python
df_trips.head()
```

```python
df_taxi_zones = pd.read_csv("data/taxi_zone_lookup.csv")
```

```python
df_taxi_zones.head()
```

```python
df_trips.info()
```

```python
df_trips = df_trips.drop(columns = "ehail_fee")
```

```python
df_trips.info()
```

```python
df_taxi_zones.info()
```

```python
int32_cols = df_trips.select_dtypes(include="int").columns
df_trips[int32_cols] = df_trips[int32_cols].astype("int64")
```

```python
df_trips.info()
```

```python
obj_cols_trips = df_trips.select_dtypes(include="object").columns
obj_cols_zones = df_taxi_zones.select_dtypes(include="object").columns
df_trips[obj_cols_trips] = df_trips[obj_cols_trips].astype("string")
df_taxi_zones[obj_cols_zones] = df_taxi_zones[obj_cols_zones].astype("string")
```
